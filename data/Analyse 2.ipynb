```python
import pandas as pd
import hashlib
import os
```

```python
def hacher_file(filePath):
    fileObj = open(filePath, 'rb')
    m = hashlib.md5()
    while True:
        d = fileObj.read(8096)
        if not d:
            break
        m.update(d)
    return m.hexdigest()
```

```python
def compare(hash1, hash_temoin):
    return (hash1 == hash_temoin)
   
```

```python
# More deep comparaison 
def deep_compare(path1, path2):
    with open(path1, 'r') as option:
        # Lire le contenu du fichier
        contenu_option = option.read().splitlines()
    with open(path2, 'r') as temoin:
        contenu_temoin = temoin.read().splitlines()
    
    for i in range(min(len(contenu_option), len(contenu_temoin))):
        if(contenu_option[i]!=contenu_temoin[i]):
            return (i, contenu_option[i], contenu_temoin[i])
    
    
    
```

```python
def nb_comb(number_comb):
    return number_comb+"comb/"
def nb_token(number_token_gen):
    if(number_token_gen == "1"):
        return "1token_gen/"
    else :
        return number_token_gen+"tokens_gen/"
```

```python
def lister_fichiers(path):
    # os.listdir() renvoie une liste contenant les noms de fichiers et dossiers
    fichiers = os.listdir(path)
    # Filtrer uniquement les fichiers
    fichiers = [f for f in fichiers]
    return fichiers
```

```python

number_comb = "1"
number_token_gen = "10"
path = nb_comb(number_comb)+nb_token(number_token_gen)+"NeuralNetwork/gpt2/"
options= lister_fichiers(path)
number = ["1.csv","2.csv","3.csv","4.csv","5.csv"]

list_path_only = []
list_path_opt = []

for i in options:
    for j in number :
        list_path_opt.append(path+i+"/input"+j)
for j in number :
    list_path_only.append(path+"only/input"+j)        
print(path)
```

```python
print("nombre de donnée save : ", len(list_path_opt))
```

```python
hash_only = []
for i in list_path_only : 
    hash_only.append(hacher_file(i))
hash_only
```

```python
like_only = []
not_like_only = []

for i in list_path_opt : 
    if("input1" in i) : 
        input_=0
    if("input2" in i) : 
        input_=1
    if("input3" in i) : 
        input_=2
    if("input4" in i) : 
        input_=3
    if("input5" in i) : 
        input_=4
    
    try:  
        hash_tmp = hacher_file(i)
        if(compare(hash_tmp, hash_only[input_])):
            like_only.append(i.split("gpt2/")[-1])
        else :
            not_like_only.append(i.split("gpt2/")[-1])
    except Exception as e:  
        print("erreur with path :", i)
```
